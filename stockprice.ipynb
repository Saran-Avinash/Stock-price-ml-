```python

import numpy as np
from skrebate import ReliefF
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
features = stock_data['TSLA'][[ 'OpenPrice', 'ClosePrice', 'HighPrice', 'LowPrice', 'Volume', 'MovingAverage10', 'MovingAverage50', 'RSI', 'MACD', 'Volatility']]
target = stock_data['TSLA']['PriceMovement'].map({'Up': 1, 'Down': 0})  # Convert 'Up' to 1, 'Down' to 0
# target

scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
print(features_scaled)

target = target.reset_index(drop=True)
```

```python
relieff = ReliefF(n_neighbors=100)
relieff.fit(features_scaled, target)

# Step 4: Get the importance of each feature
feature_importance = relieff.feature_importances_

# Step 5: Create a DataFrame to show feature weights
feature_weights = pd.DataFrame({'Feature': features.columns, 'Weight': feature_importance})

# Sort features by importance
feature_weights_sorted = feature_weights.sort_values(by='Weight', ascending=False)

# Print the sorted feature weights
print(feature_weights_sorted)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
```

```python
selected_features = stock_data['TSLA'][['OpenPrice', 'ClosePrice', 'HighPrice', 'LowPrice', 'Volume']]
target = stock_data['TSLA']['PriceMovement'].map({'Up': 1, 'Down': 0})
X_train, X_test, y_train, y_test = train_test_split(selected_features, target, test_size=0.2, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)  # Fit and transform on training data
X_test_scaled = scaler.transform(X_test)        # Only transform on test data

# Step 3: Train the Random Forest model using scaled training features
model = RandomForestClassifier(random_state=49)
model.fit(X_train_scaled, y_train)  # Use scaled training data for fitting

# Step 4: Make predictions using scaled test features
y_pred = model.predict(X_test_scaled)  # Use scaled test data for predictions

# Step 5: Evaluate the model
print(classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
